Split feature df test

```python
column_names = full_df.Name.unique().tolist()

column_names
```

```python
df_names = []

def create_df_names(list):
    for i in list:  
        i = i.replace(" ", "_")
        df_names.append(i)

create_df_names(column_names)

df_names
```

```python
#function to loop through features and pull into df 
# https://stackoverflow.com/questions/48888001/creating-multiple-dataframes-with-a-loop


dict_of_dfs ={}


#loop through each feature
for column, feature_df in zip(column_names, df_names):

    #pull those rows into df 
    dict_of_dfs[feature_df] = full_df.loc[full_df["Name"] == column] #does this need to be f""

    #rename measurement column with feature name 
    dict_of_dfs[feature_df] = dict_of_dfs[feature_df].rename(mapper={"Minute10Average":feature_df}, axis=1)

    #drop name column 
    dict_of_dfs[feature_df] = dict_of_dfs[feature_df].drop(labels=["Name"], axis=1)
```

Each feature is in a df

Each df is stored in a dictionary, where the key value pairs are df_name: df

Access each feature specific df through dict_of_dfs["df_name"], i.e. dict_of_dfs['Power']
