```python
fault_df = fault_df.rename({"Unnamed: 2":"time_stamp",
                    "Turbine_ID":"turbine_id",
                    "Unnamed: 1":"fault",
                    "Unnamed: 3":"description"}, axis=1)
```

```python
fault_df = fault_df[["time_stamp", "turbine_id", "fault", "description"]]
```

```python
fault_df
```

```python
fault_df.dtypes
```

```python
fault_df["time_stamp"] = fault_df["time_stamp"].apply(pd.to_datetime)

```

```python
fault_df.dtypes
```

```python
#round demo

fault_df["time_stamp"][0].round(freq="10T")
```

```python
#round loop
times = fault_df["time_stamp"]
round_times = []
for i in times:
    round_times.append(i.round(freq="10T").replace(tzinfo=None))
    
```

```python
#rounded time to df 
fault_df["time_stamp"] = round_times
```
