* task: pull list of tuples for each turbine into dataframes 
* issue: type is string, not list; need to convert w json? => resolved, used ast
* task: tuples to dict for df => resolved, function convert()

```python
#backup to json loads bcs two columns 
import ast 
```

```python
#index col = 0 to use unique index, rather than push to column 

feature_df = pd.read_csv("Dataframes/feature_importances.csv", index_col=0)
```

```python
feature_df
```

```python
# pull row info and turn into list of tuples 
T11 = (feature_df.iat[0,0])

T11_features = ast.literal_eval(T11)
```

```python
T06 = (feature_df.iat[1,0])

T06_features = ast.literal_eval(T06)
```

```python
T07 = (feature_df.iat[2,0])

T07_features = ast.literal_eval(T07)
```

```python
T01 = (feature_df.iat[3,0])

T01_features = ast.literal_eval(T01)
```

convert all tuples to dicts for df 

```python
def Convert(tup, di):
    for a, b in tup:
        di.setdefault(b, a)
    return di

tups_T11 = T11_features
dict_T11 = {}

tups_T06 = T06_features
dict_T06 = {}

tups_T07 = T07_features
dict_T07 = {}

tups_T01 = T01_features
dict_T01 = {}

Convert(tups_T11, dict_T11)

Convert(tups_T06, dict_T06)

Convert(tups_T07, dict_T07)

Convert(tups_T01, dict_T01)
```

```python
df = pd.DataFrame(data=[dict_T01, dict_T06, dict_T07, dict_T11], index=["T01", "T06", "T07", "T11"])
```
